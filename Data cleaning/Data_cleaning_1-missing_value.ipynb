.2)Fill missing value manually

.3)Global constant->Constant e.g 0 or unknown

.4)Measure of central tendency(Mean,Midian,Mode)-> fill with central tendency

```python
nv = dict(df.isnull().sum()/df.shape[0]*100) #make through dictionary making
drop_columns = [i for i in nv.keys() if nv[i] > 17]
drop_columns
```

```python
null_var = df.isnull().sum()/df.shape[0]*100
dc = null_var[null_var>17].keys() #direct method
dc
```

```python
df2 = df.drop(columns = drop_columns)
df2.shape
```

```python
plt.figure(figsize = (25,25)) #tell the figure size
sns.heatmap(df2.isnull()) #less white space
```

```python
df3 = df2.dropna() #bydefault row wise deletion if any none value
df3.shape
```

```python
pd.concat([df['MSZoning'].value_counts()/df.shape[0]*100 ,
           df3['MSZoning'].value_counts()/df.shape[0]*100 ],axis =1  #column wise column
          ,keys = ['MSZoning_org','MSZoning_after_clean'])  #rename the column

          #total loss after cleaning the data
```

```python
def cat_var_dist(var):
  return pd.concat([df[var].value_counts()/df.shape[0]*100 ,
           df3[var].value_counts()/df.shape[0]*100 ],axis =1  #column wise column
          ,keys = [var+'_org',var+'_after_clean'])
for i in num_var:
  print(cat_var_dist(i))
  print('\n')
```
