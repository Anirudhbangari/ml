>Major steps in Data Preprocessing


.Data Cleaning-->Means fill in missing values,smooth out noise while identifying outlier, and correct inconsistencies in the data


.Data Integration-->It is techniques to merges data from multiple sources into a coherent data store,such as a data warehouse.


.Data Reduction-->It is a technique use to reduce the data size by aggregating,eliminating redundant features,or clusterin,for instance.


.Data Transformation-->It means data are transformed or consolidated into forms apporipate for ML model traning,such as normalization,may be applied where data are scaled to fall within a smaller range like 0 to 1
