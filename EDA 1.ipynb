```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('heart_failure_clinical_records_dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

#1. Distribution of Age among Heart Failure Patients


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of Age among Heart Failure Patients
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Distribution of Age among Heart Failure Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

#2. Death Rate Variation with Age

```python
# Death Rate Variation with Age
plt.figure(figsize=(10, 6))
sns.histplot(df[df['DEATH_EVENT'] == 1]['age'], bins=20, kde=True, color='red', label='Death Event')
sns.histplot(df[df['DEATH_EVENT'] == 0]['age'], bins=20, kde=True, color='green', label='Survived')
plt.title('Death Rate Variation with Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

#3. Percentage of Male and Female Patients

```python
# Percentage of Male and Female Patients
gender_counts = df['sex'].value_counts(normalize=True) * 100
plt.figure(figsize=(8, 6))
sns.barplot(x=gender_counts.index, y=gender_counts.values, palette='coolwarm')
plt.title('Percentage of Male and Female Patients')
plt.xlabel('Gender (0 = Female, 1 = Male)')
plt.ylabel('Percentage')
plt.show()
```

#4. Platelet Count Variation among Different Age Groups

```python
# Platelet Count Variation among Different Age Groups
plt.figure(figsize=(10, 6))
sns.boxplot(x=pd.cut(df['age'], bins=[40, 50, 60, 70, 80, 95]), y='platelets', data=df)
plt.title('Platelet Count Variation among Different Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Platelet Count')
plt.show()
```

#5. Correlation between Creatinine and Sodium Levels

```python
# Correlation between Creatinine and Sodium Levels
plt.figure(figsize=(10, 6))
sns.scatterplot(x='serum_creatinine', y='serum_sodium', data=df, hue='DEATH_EVENT', palette='coolwarm')
plt.title('Correlation between Creatinine and Sodium Levels')
plt.xlabel('Serum Creatinine')
plt.ylabel('Serum Sodium')
plt.show()
```

#6. Prevalence of High Blood Pressure between Male and Female Patients

```python
# Prevalence of High Blood Pressure between Male and Female Patients
plt.figure(figsize=(10, 6))
sns.countplot(x='sex', hue='high_blood_pressure', data=df, palette='Set2')
plt.title('Prevalence of High Blood Pressure between Male and Female Patients')
plt.xlabel('Gender (0 = Female, 1 = Male)')
plt.ylabel('Count')
plt.show()
```

#7. Relationship between Smoking Habits and Heart Failure

```python
# Relationship between Smoking Habits and Heart Failure
plt.figure(figsize=(10, 6))
sns.countplot(x='smoking', hue='DEATH_EVENT', data=df, palette='Set1')
plt.title('Relationship between Smoking Habits and Heart Failure')
plt.xlabel('Smoking (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

#8.Patterns in the Distribution of Death Events across Age Groups


```python
# Patterns in the Distribution of Death Events across Age Groups
plt.figure(figsize=(10, 6))
sns.histplot(df[df['DEATH_EVENT'] == 1]['age'], bins=20, kde=True, color='red', label='Death Event')
plt.title('Distribution of Death Events across Age Groups')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

#9.Difference in Ejection Fraction between Patients with and without Diabetes

```python
# Difference in Ejection Fraction between Patients with and without Diabetes
plt.figure(figsize=(10, 6))
sns.boxplot(x='diabetes', y='ejection_fraction', data=df, palette='Set3')
plt.title('Difference in Ejection Fraction between Patients with and without Diabetes')
plt.xlabel('Diabetes (0 = No, 1 = Yes)')
plt.ylabel('Ejection Fraction')
plt.show()
```

#10. Serum Creatinine Level Variation between Patients Who Survived and Those Who Did Not


```python
# Serum Creatinine Level Variation between Patients Who Survived and Those Who Did Not
plt.figure(figsize=(10, 6))
sns.boxplot(x='DEATH_EVENT', y='serum_creatinine', data=df, palette='Set2')
plt.title('Serum Creatinine Level Variation between Survived and Not Survived Patients')
plt.xlabel('Death Event (0 = No, 1 = Yes)')
plt.ylabel('Serum Creatinine')
plt.show()
```
